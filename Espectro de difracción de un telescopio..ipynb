$ \begin{equation}
J_{m}(x)=\frac{1}{\pi} \int_{0}^{\pi} cos(m\theta-xsin\theta) \cdot d\theta
\end{equation}$

```python
from numpy import *
from pylab import *
import matplotlib.pyplot as plt
for m in range(0,3):
    def J(m,x):
        def f(teta):
            return cos(m*teta - x*sin(teta))
    
        N = 1000
        a = 0.
        b = pi
        h = (b-a)/N
        s = f(a) + f(b) + 4*f(b-h)
        for k in range(0,N//2):
            s += 4*f(a + (2*k-1)*h) + 2*f(a+2*k*h)
    
        i = h/3*s/pi
        print("Valor m=",m,"\n",i) 
        return i
        
x = linspace(0,20)
plt.title("Intregal para direntes m:")
plot(x,J(0,x),label='J0')
plot(x,J(1,x),label='J1')
plot(x,J(2,x),label='J2')
plt.legend()
```
